# Basilic Data preparation - French cultural venues references

**Project: Analysis of cultural accessibility and territorial inequalities in France**

### Research Questions:
1. Geographic Distribution: How is cultural supply distributed across French territories?
2. Urban vs Rural: Are there significant disparities between urban and rural areas?
3. Socio-Economic Correlation: Is there a relationship between territorial wealth and cultural supply?
4. Typological Diversity: What types of cultural venues exist and how are they distributed?
5. PACA Regional Focus: How does PACA compare to national averages?

## Dataset Information

**Source:** French Ministry of Culture (Ministère de la Culture)

**Name:** BASILIC - Database of Cultural Equipment and Venues in France

**Origin:** Official government database aggregating cultural infrastructure data from multiple ministerial sources

**Year:** 2025 (snapshot from September 5, 2025)

**Last Update:** September 5, 2025

**Content:** Comprehensive inventory of cultural venues and equipment across France, including cinemas, theaters, libraries, museums, archives, and other cultural facilities (~30,000+ venues)

**File Location:** `data/raw/basilic-dataculture-2025-09-05.xlsx`

**Format:** Excel (.xlsx)

**Purpose:** Reference database for analyzing cultural supply, territorial inequalities, and cultural accessibility across French territories

---

## Key Information

**Geographic Coverage:** All French territories (metropolitan France and overseas territories)

**Types of Cultural Venues:**
- Cinemas (cinémas)
- Theaters (théâtres)
- Libraries (bibliothèques)
- Archive services (services d'archives)
- Museums and cultural institutions
- Concert halls and performance spaces
- Cultural centers

**Data Quality:**
- Complete geographic coordinates (latitude/longitude) for mapping
- INSEE codes for administrative linking
- Detailed typology of equipment
- Socio-economic indicators (priority neighborhoods/QPV)

---

## Choice of columns

**Must have (geographic and administrative) :**
- `id`: Unique identifier (auto-generated)
- `nom`: Name of the cultural venue
- `latitude`: Latitude coordinate (WGS84)
- `longitude`: Longitude coordinate (WGS84)
- `code_insee`: INSEE commune code (5 digits)
- `code_postal`: Postal code (5 digits)
- `region`: Region name
- `type_equipement_ou_lieu`: Type of equipment/venue

**Socio-Economic analysis:**
- `departement`: Department name
- `dept_code`: Department code (2 digits)
- `domaine`: Main cultural domain
- `sous_domaine`: Sub-domain/category
- `identifiant_qpv_24`: Priority neighborhood identifier (Quartiers Prioritaires de la Politique de la Ville)
- `label_et_appellation`: Official labels and designations

**Venue specific details:**

*Cinema:*
- `nombre_fauteuils_de_cinema`: Number of cinema seats
- `nombre_ecrans`: Number of screens
- `multiplexe`: Multiplex flag
- `type_de_cinema`: Cinema type

*Theater:*
- `nombre_de_salles_de_theatre`: Number of theater halls
- `jauge_du_theatre`: Theater capacity

*Library:*
- `surface_bibliotheque`: Library surface area (m²)
- `nom_du_reseau_de_bibliotheques`: Library network name

*Additional:*
- `adresse`: Street address
- `annee_label_appellation`: Year of label/designation
- `precision_equipement`: Equipment precision/details

## Cleaning and adaptation of the database

```python
print("Remove duplicate rows")
print()


before = len(df_culture)
df_culture = df_culture.drop_duplicates()
after = len(df_culture)
removed = before - after

print(f"Before: {before:,} rows")
print(f"After:  {after:,} rows")
print(f"Removed {removed:,} duplicates")
```

```python
print("Clean postal codes")
print()


df_culture['Code_Postal'] = df_culture['Code_Postal'].astype(str)
df_culture['Code_Postal'] = df_culture['Code_Postal'].str.replace('.0', '', regex=False)
df_culture['Code_Postal'] = df_culture['Code_Postal'].str.zfill(5)
df_culture['dept_code'] = df_culture['Code_Postal'].str[:2]


print(f"Examples: {df_culture['Code_Postal'].head(5).tolist()}")

```

```python
print("Clean INSEE code")
print()

if 'code_insee' in df_culture.columns:
    df_culture['code_insee'] = df_culture['code_insee'].astype(str)
    df_culture['code_insee'] = df_culture['code_insee'].str.replace('.0', '', regex=False)
    df_culture['code_insee'] = df_culture['code_insee'].str.zfill(5)
    print(f"✅ INSEE codes cleaned")
    print(f"Examples: {df_culture['code_insee'].head(5).tolist()}")
else:
    insee_cols = [col for col in df_culture.columns if 'insee' in col.lower()]
    print(f"NSEE column: {insee_cols}")
```

```python
print("Clean numeric columns")
print()

numeric_cols = [
    'Nombre_fauteuils_de_cinema',
    'Nombre_ecrans',
    'Nombre_de_salles_de_theatre',
    'Jauge_du_theatre',
    'Surface_Bibliotheque'
]

for col in numeric_cols:
    if col in df_culture.columns:
        df_culture[col] = pd.to_numeric(df_culture[col], errors='coerce')  # to not crash
        non_null = df_culture[col].notna().sum()
        print(f" {col:35s}: {non_null:,} values")





```

```python
print("Clean text columns")
print()

text_cols = [
    'Nom', 
    'Type_equipement_ou_lieu', 
    'Region', 
    'Departement',
    'Domaine',
    'Sous_domaine',
    'Label_et_appellation'
]

for col in text_cols:
    if col in df_culture.columns:
        df_culture[col] = df_culture[col].str.strip()
        print(f"Cleaned: {col}")
```

```python
print("Fix column names for SQL")
print()

def make_sql_friendly(column_name):   #Convert column names to SQL-friendly format
    name = column_name.lower()
    name = name.replace(' ', '_')
    name = name.replace('-', '_')
    
    # Remove French accents
    accents = {
        'é': 'e', 'è': 'e', 'ê': 'e', 'ë': 'e',
        'à': 'a', 'â': 'a', 'ä': 'a',
        'ô': 'o', 'ö': 'o',
        'û': 'u', 'ù': 'u', 'ü': 'u',
        'ç': 'c',
        'î': 'i', 'ï': 'i'
    }
    for accent, replacement in accents.items():
        name = name.replace(accent, replacement)
    
    name = name.replace("'", '')
    name = name.replace('(', '')
    name = name.replace(')', '')
    name = name.replace('.', '')
    name = name.strip('_')
    
    return name

old_columns = df_culture.columns.tolist()
df_culture.columns = [make_sql_friendly(col) for col in df_culture.columns]
new_columns = df_culture.columns.tolist()

print("Column names converted to SQL-friendly format")
print("\nExamples of changes:")
for i in range(min(10, len(old_columns))):
    if old_columns[i] != new_columns[i]:
        print(f"   {old_columns[i]} → {new_columns[i]}")

```

```python
print("NaN (missing) values")
print()

# Fill text columns with "Unknown"
text_fill_cols = [
    'type_equipement_ou_lieu',
    'region',
    'departement',
    'domaine',
    'sous_domaine',
    'label_et_appellation'
]

print("✅ Filling NaN in text columns with 'Unknown':")
for col in text_fill_cols:
    if col in df_culture.columns:
        nan_count = df_culture[col].isna().sum()
        if nan_count > 0:
            df_culture[col] = df_culture[col].fillna('Unknown')
            print(f"   {col:40s}: filled {nan_count:,} NaN")

# Numeric columns keep NaN (will be NULL in SQL)
print("\n Numeric columns will keep NaN (!! becomes NULL in SQL !!)")
```

```python
print("Standardize text case")


title_case_cols = ['nom', 'region', 'departement', 'type_equipement_ou_lieu', 'domaine']

for col in title_case_cols:
    if col in df_culture.columns:
        df_culture[col] = df_culture[col].str.title()
        examples = df_culture[col].value_counts().head(3).index.tolist()
        print(f" {col:}: {examples}")

# Uppercase for codes
code_cols = ['code_postal', 'code_insee', 'dept_code']
for col in code_cols:
    if col in df_culture.columns:
        df_culture[col] = df_culture[col].astype(str).str.upper()
```

## Preparation for SQL : 24 columns selected

### 1. Must have (MH)
### 2. Socio-economical analysis (SEA)
### 3. Venues specifics (VS)

```python
# Must have --> MH

df_cultural_MH = [
    'id',
    'nom',
    'latitude',
    'longitude',
    'code_insee',
    'code_postal',
    'region',
    'type_equipement_ou_lieu',
]


# Socio-economical analysis --> SEA

df_cultural_SEA = [
    'departement',
    'dept_code',
    'domaine',
    'sous_domaine',
    'identifiant_qpv_24',      # Priority neighborhoods : linked with public subventions
    'label_et_appellation',
]

# Venues specifics --> VS

df_cultural_VS = [
    # Cinema
    'nombre_fauteuils_de_cinema',
    'nombre_ecrans',
    'multiplexe',
    'type_de_cinema',
    
    # Theater
    'nombre_de_salles_de_theatre',
    'jauge_du_theatre',
    
    # Library
    'surface_bibliotheque',
    'nom_du_reseau_de_bibliotheques',
    
    # Additional
    'adresse',
    'annee_label_appellation',
    'precision_equipement',
]


```

```python

# Concept

df_cultural_sql = df_cultural_MH + df_cultural_SEA + df_cultural_VS


existing_columns = [col for col in df_cultural_sql if col in df_culture.columns]

# Show what we have
print(f"\nMust have ({len(df_cultural_MH)} columns):")
for col in df_cultural_MH:
    status = "" if col in df_culture.columns else "Missing"
    print(f"   {status} {col}")

print(f"\nSocio economical analysis ({len(df_cultural_SEA)} columns):")
for col in df_cultural_SEA:
    status =  "" if col in df_culture.columns else "Missing"
    print(f"   {status} {col}")

print(f"\nVery specifics ({len(df_cultural_VS)} columns):")
for col in df_cultural_VS:
    status =  "" if col in df_culture.columns else "Missing"
    print(f"   {status} {col}")

df_cultural_VS_found = [col for col in df_cultural_VS if col in df_culture.columns]
print(f"   Found {len(df_cultural_VS_found)}/{len(df_cultural_VS)} optional columns")
```

```python
# Create final dataframe
df_cultural_final = df_culture[existing_columns].copy()

print(f"\n FINAL: {len(existing_columns)} columns selected")
```

```python
# Save CSV

import os
os.makedirs('data/processed', exist_ok=True)

output_file = 'data/processed/basilic_for_sql.csv'
df_cultural_final.to_csv(output_file, index=False, encoding='utf-8')

print(f"Saved to: {output_file}")
print(f"   Rows: {len(df_cultural_final):,}")
print(f"   Columns: {len(df_cultural_final.columns)}")

print()

print(f" TOTAL: {len(df_cultural_final.columns)} columns""")
```
