## Dataset Joining

**Strategy:** Join cultural venues with population and income data at the commune level.

**Join key:** `code_insee` (commune code - 5 digits)

**Why LEFT JOIN:** 
- Keep all cultural venues (even if no population related to income data)
- Population/Income values are missing for small communes 

```python
# MErge 1 -> culture + communes (population) 

# Drop dept_code from culture to avoid duplicates

if 'dept_code' in df_culture.columns:
    df_culture_clean = df_culture.drop(columns=['dept_code'])
    print("   Dropped dept_code from culture (will use from communes)")
else:
    df_culture_clean = df_culture.copy()
                                           #Dropped dept_code from culture (will use from communes)


# Merge
df = df_culture_clean.merge(
    df_population,
    left_on='code_insee',
    right_on='commune_code',
    how='left')

print(f"Rows: {len(df):,}")
print(f"Columns: {len(df.columns)}")
print(f"With population: {df['population'].notna().sum():,} ({df['population'].notna().mean()*100:.1f}%)")

#Rows: 88,037
#Columns: 30
#With population: 83,088 (94.4%)
```

```python
# Merge 2 : adding income


df = df.merge(
    df_income,
    on='code_insee',
    how='left')

print(f"\nRows: {len(df):,}")
print(f"Columns: {len(df.columns)}")
print(f"With income: {df['median_income'].notna().sum():,} ({df['median_income'].notna().mean()*100:.1f}%)")

#Rows: 88,037
#Columns: 39
#With income: 38,798 (44.1%)
```

### Research question 2 : Urban vs Rural
#### Are there disparities in cultural access between urban and rural areas?

```python
# Create urban/rural categories based on population

def categorize_urban(pop):        # Bins of population, by quantity
    if pd.isna(pop):
        return 'Unknown'
    elif pop < 2000:
        return 'Rural (<2k)'
    elif pop < 10000:
        return 'Small Urban (2k-10k)'
    elif pop < 50000:
        return 'Medium Urban (10k-50k)'
    else:
        return 'Large Urban (>50k)'

df['urban_category'] = df['population'].apply(categorize_urban)


urban_stats = df.groupby('urban_category').agg({
    'id': 'count',
    'code_insee': 'nunique',
    'population': 'sum'
}).rename(columns={'id': 'venues', 'code_insee': 'communes'})

urban_stats['venues_per_commune'] = urban_stats['venues'] / urban_stats['communes']
urban_stats['density_per_100k'] = (urban_stats['venues'] / urban_stats['population']) * 100000

print(f"\nDistribution by urban category : {urban_stats}")



        
```

```python
# Aggregate statistics, by category

categories = ['Rural (<2k)', 'Small Urban (2k-10k)', 'Medium Urban (10k-50k)', 'Large Urban (>50k)']


aggregate_stats = []

for cat in categories:
    cat_data = df[df['urban_category'] == cat]
    
    # Remove rows without population
    cat_data_clean = cat_data[cat_data['population'].notna()]
    cat_data_clean = cat_data_clean[cat_data_clean['population'] > 0]
    
    if len(cat_data_clean) > 0:                        #  AGGREGATE: Total venues / Total population
        total_venues = len(cat_data_clean)
        total_population = cat_data_clean['population'].sum()
        aggregate_density = (total_venues / total_population) * 100000
        
        
        n_communes = cat_data_clean['code_insee'].nunique()               # Number of communes
        
        aggregate_stats.append({
            'category': cat,
            'n_communes': n_communes,
            'total_venues': total_venues,
            'total_population': total_population,
            'aggregate_density': aggregate_density})
        
        print(f"\n{cat}:")
        print(f"   Communes: {n_communes:,}")
        print(f"   Total venues: {total_venues:,}")
        print(f"   Total population: {total_population:,.0f}")
        print(f"   Aggregate density: {aggregate_density:.1f} per 100k")
    else:
        print(f"\n{cat}: No data")



```

```python
# Comparaison

if len(aggregate_stats) >= 2:
    rural_stats = next((s for s in aggregate_stats if 'Rural' in s['category']), None)           # Get rural and large urban
    large_stats = next((s for s in aggregate_stats if 'Large' in s['category']), None)
    
    if rural_stats and large_stats:
        ratio = large_stats['aggregate_density'] / rural_stats['aggregate_density']
        
        print(f"\nRural areas:")
        print(f"   Density: {rural_stats['aggregate_density']:.1f} per 100k")
        
        print(f"\nLarge Urban areas:")
        print(f"   Density: {large_stats['aggregate_density']:.1f} per 100k")
        
        print(f"\nRatio: {ratio:.2f}x")
        if ratio > 1:
            print(f"→ Large Urban has {ratio:.1f}x more cultural venues than Rural")
        else:
            print(f"→ Rural has {1/ratio:.1f}x more cultural venues than Large Urban (unexpected!)")

```

```python
# Boxplot (distribution analysis)


data_for_box = []            # dictionnary

for cat in categories:
    cat_data = df[df['urban_category'] == cat]
    
    
    commune_density = cat_data.groupby('code_insee').agg({
        'id': 'count',
        'population': 'first'})                                 # Calculate density by commune
    
    commune_density = commune_density[commune_density['population'].notna()]
    commune_density = commune_density[commune_density['population'] > 0]
    commune_density['density'] = (commune_density['id'] / commune_density['population']) * 100000
    
    density_values = commune_density['density'].replace([np.inf, -np.inf], np.nan).dropna()
    
    data_for_box.append(density_values)
    
    if len(density_values) > 0:
        print(f"\n{cat}:")
        print(f"   Median commune density: {density_values.median():.1f} per 100k")
        print(f"   Mean commune density: {density_values.mean():.1f} per 100k")
        

# Note: Mean can be inflated by small communes

```

```python
# Plot boxplot  (variation between communes)

plt.figure(figsize=(12, 6))
plt.boxplot(data_for_box, labels=categories)
plt.ylabel('Cultural Density (venues per 100k)', fontweight='bold')
plt.xlabel('Urban Category', fontweight='bold')
plt.title('Distribution of Cultural Density by Urban Category\n(Per Commune - shows variation)', 
          fontweight='bold', fontsize=14)
plt.xticks(rotation=15, ha='right')
plt.grid(axis='y', alpha=0.3)
plt.tight_layout()
plt.show()
```

```python
# ANOVA on commune densities (variation between communes)


valid_data = [d for d in data_for_box if len(d) > 0]

if len(valid_data) >= 2:
    f_stat, p_value_anova = st.f_oneway(*valid_data)
    
    print(f"\nH0: No difference in commune-level densities across categories")
    print(f"H1: Significant differences exist")
    print(f"\nF-statistic: {f_stat:.2f}")
    print(f"P-value: {p_value_anova:.10f}")
    
    if p_value_anova < 0.05:
        print(f"\nReject H0 (p < 0.05)")
        print(f"→ Commune-level densities differ significantly between categories")
      
    else:
        print(f"\nFail to reject H0")
else:
    print("\n Not enough groups for ANOVA")            # at least 2

# Note: Mean can be inflated by small communes : BIAS
```

```python
# It is better to use aggregate density (total venues / total population) when comparing categories of different sizes,
# as it reflects the true overall cultural access at the category level. 
# The boxplot and ANOVA show the variation between individual communes within each category, 
# not the true aggregate density across the entire category.
# For exemple: small communes inflate the average

## Formula : 
# Aggregate Density = (Total Venues / Total Population) × 100,000
```

#### Calculate aggregate density 

```python

# Define categories in order
categories = ['Rural (<2k)', 'Small Urban (2k-10k)', 
              'Medium Urban (10k-50k)', 'Large Urban (>50k)']


# Store results
results = []                                            


for category in categories:
    
    
    category_data = df[df['urban_category'] == category]
    
  
    category_clean = category_data[category_data['population'].notna()]
    category_clean = category_clean[category_clean['population'] > 0]
    
    if len(category_clean) > 0:
       
        total_venues = len(category_clean)
        total_population = category_clean['population'].sum()
        aggregate_density = (total_venues / total_population) * 100000
        unique_communes = category_clean['code_insee'].nunique()
        result_dict = {
            'category': category,
            'communes': unique_communes,
            'venues': total_venues,
            'population': total_population,
            'density': aggregate_density}
        results.append(result_dict)
        
        
        print(f"\n{category}:")
        print(f"   Communes: {unique_communes:,}")
        print(f"   Total venues: {total_venues:,}")
        print(f"   Total population: {total_population:,.0f}")
        print(f"   Aggregate density: {aggregate_density:.1f} per 100k inhabitants")
```

```python
# Compare rural vs LArge urban 

# Find rural stats
rural = next((r for r in results if 'Rural' in r['category']), None)

# Find large urban stats
large_urban = next((r for r in results if 'Large' in r['category']), None)

if rural and large_urban:
    
    # Calculate ratio
    ratio = large_urban['density'] / rural['density']
    
    print(f"\nRural areas (<2k inhabitants):")
    print(f"   Density: {rural['density']:.1f} per 100k")
    
    print(f"\nLarge Urban areas (>50k inhabitants):")
    print(f"   Density: {large_urban['density']:.1f} per 100k")
    
    print(f"\n Ratio:")
    if ratio > 1:
        print(f"   Large Urban has {ratio:.2f}x MORE cultural venues than Rural")
        print(f"   ({(ratio-1)*100:.0f}% higher)")
    else:
        print(f"   Large Urban has {ratio:.2f}x LESS cultural venues than Rural")
        print(f"   ({(1-ratio)*100:.0f}% lower)")
```

```python
# Bar chart (Visualisation)


categories_plot = [r['category'] for r in results]
densities_plot = [r['density'] for r in results]


plt.figure(figsize=(12, 6))
bars = plt.bar(range(len(categories_plot)), densities_plot, 
               color='#26A69A', edgecolor='#004D40', linewidth=1.5)


bars[0].set_color('#80CBC4')  # Rural - lighter
bars[-1].set_color('#004D40')  # Large urban - darker

# Add value labels on bars
for i, (cat, density) in enumerate(zip(categories_plot, densities_plot)):
    plt.text(i, density + 5, f'{density:.1f}', 
             ha='center', va='bottom', fontweight='bold')

# Labels and title
plt.xlabel('Urban Category', fontweight='bold', fontsize=12)
plt.ylabel('Cultural Density (venues per 100k inhabitants)', fontweight='bold', fontsize=12)
plt.title('Cultural Density by Urban Category (Aggregate Method)', 
          fontweight='bold', fontsize=14)
plt.xticks(range(len(categories_plot)), categories_plot, rotation=15, ha='right')
plt.grid(axis='y', alpha=0.3)
plt.tight_layout()
plt.show()
```

```python
# Chi Square (statistical test) 

print("\nChi Square - statistical test")
print("\nHypotheses:")
print("   H0: Cultural venues are uniformly distributed across urban categories")
print("   H1: Cultural venues are not uniformly distributed")

# Prepare data for chi-square
observed = [r['venues'] for r in results]
total_venues_all = sum(observed)

# Expected: if venues were proportional to population
total_population_all = sum([r['population'] for r in results])
expected = [(r['population'] / total_population_all) * total_venues_all 
            for r in results]

# Perform chi-square test
chi2, p_value = st.chisquare(observed, expected)

print(f"\n Test Results:")
print(f"   Chi-square statistic: {chi2:.2f}")
print(f"   P-value: {p_value:.10f}")
print(f"   Significance level: α = 0.05")

if p_value < 0.05:
    print(f"\n   REJECT H0 (p < 0.05)")
else:
    print(f"\n   FAIL TO REJECT H0 (p >= 0.05)")
    
```

#### Chi Square test :
Cultural venues are not uniformly distributed :
The chi-square test confirms that cultural venues are not uniformly distributed across 
Urban categories (χ² = 1,387,458, p < 0.001). 

This indicates significant disparities in cultural access between urban and rural areas, 
with the distribution heavily concentrated in urban centers.


#### Key Findings :
Aggregate density shows the overall cultural access for each category, for the total venues available to the total population.

   • Unexpected: Rural areas have higher density
   ->This may indicate different patterns of cultural distribution

   • Statistical test confirms significant differences (p < 0.001)
    -> Clear urban-rural divide exists in cultural accessibility




### Research question 3 : Socio-economic correlation
#### Is there a relationship between territorial wealth and cultural supply?


```python
# Calculate cultural density by commune
commune_stats = df.groupby('code_insee').agg({
    'id': 'count',
    'population': 'first',
    'median_income': 'first',
    'poverty_rate': 'first'
}).rename(columns={'id': 'venue_count'})

commune_stats['density_per_100k'] = (commune_stats['venue_count'] / commune_stats['population']) * 100000

# Filter valid data (both income and density)
valid_communes = commune_stats[(commune_stats['median_income'].notna()) & (commune_stats['density_per_100k'].notna())]

print(f"\n Communes with both income and cultural data: {len(valid_communes):,}")

# Visualization: Scatter plot
plt.figure(figsize=(12, 6))
plt.scatter(valid_communes['median_income'], 
            valid_communes['density_per_100k'],
            alpha=0.5, 
            color='#80CBC4',
            edgecolors='#004D40',
            s=50)

# Add trend line
z = np.polyfit(valid_communes['median_income'], valid_communes['density_per_100k'], 1)
p = np.poly1d(z)
plt.plot(valid_communes['median_income'].sort_values(), 
         p(valid_communes['median_income'].sort_values()), 
         "r--", 
         linewidth=2,
         label='Trend line')

plt.xlabel('Median Income (€)', fontweight='bold')
plt.ylabel('Cultural Density (venues per 100k)', fontweight='bold')
plt.title('Correlation: Income vs Cultural Density', fontweight='bold', fontsize=14)
plt.legend()
plt.grid(alpha=0.3)
plt.tight_layout()
plt.show()

```

```python
# Statistical test: Pearson correlation
# H0: No correlation between income and cultural density (ρ = 0)
# H1: Significant correlation exists (ρ != 0)

correlation, p_value_corr = st.pearsonr(valid_communes['median_income'], 
                                         valid_communes['density_per_100k'])

print("\n Pearson Correlation (statistical test)")
print(f"   H0: No correlation between income and cultural density (ρ = 0)")
print(f"   H1: Significant correlation exists (ρ ≠ 0)")
print(f"\n   Correlation coefficient (r): {correlation:.4f}")
print(f"   P-value: {p_value_corr:.10f}")
print(f"   Sample size: {len(valid_communes):,}")

if p_value_corr < 0.05:
    print(f"\n   REJECT H0 (p < 0.05)")
    if correlation > 0:
        strength = "weak" if abs(correlation) < 0.3 else "moderate" if abs(correlation) < 0.7 else "strong"
        print(f"   -> There is a {strength} positive correlation ({correlation:.3f})")
        print(f"      Higher income communes tend to have more cultural venues")
    else:
        print(f"   -> There is a negative correlation")
else:
    print(f"\n Fail to reject H0 (p >= 0.05)")

```

```python
# Poverty correlation
valid_poverty = commune_stats[(commune_stats['poverty_rate'].notna()) & (commune_stats['density_per_100k'].notna())]

if len(valid_poverty) > 30:
    corr_poverty, p_poverty = st.pearsonr(valid_poverty['poverty_rate'], valid_poverty['density_per_100k'])
    print(f"\n Poverty rate correlation")
    print(f"   Correlation with cultural density: {corr_poverty:.4f}")
    print(f"   P-value: {p_poverty:.6f}")
```

### Research question 4 : Cultural diversity
#### What types exist and how is the diversity distributed?



```python
# How diverse is the cultural offering across territories?

# Observation of departments (diversity)

# Calculate diversity by department
dept_diversity = df.groupby('dept_code').agg({
    'type_equipement_ou_lieu': 'nunique',           # Number of unique types
    'id': 'count'}).rename(columns={
    'type_equipement_ou_lieu': 'unique_types',
    'id': 'total_venues'})

# Diversity index (types per 100 venues)
dept_diversity['diversity_index'] = (dept_diversity['unique_types'] / dept_diversity['total_venues']) * 100

dept_diversity = dept_diversity.sort_values('unique_types', ascending=False)

print(f"\nOverall statistics")
print(f"   Total unique venue types in France: {df['type_equipement_ou_lieu'].nunique()}")
print(f"   Average types per department: {dept_diversity['unique_types'].mean():1f}")
print(f"   Median types per department: {dept_diversity['unique_types'].median():1f}")

print(f"\nTop 10 department (in terms of diversity")
for i, (dept, row) in enumerate(dept_diversity.head(10).iterrows(), 1):
    print(f"{i:2f}. Dept {dept}: {row['unique_types']:3f} types, {row['total_venues']:5,} venues (index: {row['diversity_index']:.1f})")

print(f"\n10 least diverse departments")
for i, (dept, row) in enumerate(dept_diversity.tail(10).iterrows(), 1):
    print(f"{i:2f}. Dept {dept}: {row['unique_types']:3f} types, {row['total_venues']:5,} venues (index: {row['diversity_index']:.1f})")



```

```python
# Visualization 1: diversity vs volume (scatter plot)

fig1, ax1 = plt.subplots(figsize=(12, 6))

ax1.scatter(dept_diversity['total_venues'], 
            dept_diversity['unique_types'],
            alpha=0.6, 
            s=100,
            color='#26A69A',
            edgecolors='#004D40')

# trend line
z = np.polyfit(dept_diversity['total_venues'], dept_diversity['unique_types'], 1)
p = np.poly1d(z)
x_line = np.linspace(dept_diversity['total_venues'].min(), dept_diversity['total_venues'].max(), 100)
ax1.plot(x_line, p(x_line), '--', linewidth=2, color='#FF6F61', label='trend')

ax1.set_xlabel('total number of venues', fontweight='bold')
ax1.set_ylabel('number of unique venue types', fontweight='bold')
ax1.set_title('cultural diversity vs volume', fontweight='bold', fontsize=12)
ax1.legend()
ax1.grid(alpha=0.3)

plt.tight_layout()
plt.show()
```

```python
# visualization 2: top 15 most diverse departments (bar chart)
fig2, ax2 = plt.subplots(figsize=(12, 8))

top15_div = dept_diversity.head(15)
ax2.barh(range(len(top15_div)), top15_div['unique_types'], color='#26A69A')
ax2.set_yticks(range(len(top15_div)))
ax2.set_yticklabels(top15_div.index)
ax2.set_xlabel('number of unique types', fontweight='bold')
ax2.set_title('top 15 most diverse departments', fontweight='bold', fontsize=12)
ax2.invert_yaxis()

for i, v in enumerate(top15_div['unique_types']):
    ax2.text(v + 0.5, i, f'{int(v)}', va='center')

plt.tight_layout()
plt.show()                                        # integer
```

```python
# Correlation test
corr_div, p_div = st.pearsonr(dept_diversity['total_venues'], dept_diversity['unique_types'])

print(f"\nCorrelation: Total Venues vs Diversity")
print(f"Correlation coefficient: {corr_div:.4f}")
print(f"P-value: {p_div:.10f}")

if p_div < 0.05:
    print("Significant correlation")
else:
    print("No significant correlation")

#  More venues leads here to more diversity of venues
```

#### Key findings : 
- The analysis reveals a clear diversity-volume relationship: territories with larger numbers of cultural venues systematically offer more diverse cultural experiences. 
- This suggests that cultural diversity is not randomly distributed but follows patterns of urban development and population density.
- Metropolitan departments show relatively homogeneous diversity (14-20 types), while overseas territories exhibit significantly lower diversity (1-11 types), likely due to smaller populations, geographic isolation, and different cultural infrastructure priorities.

#### Geographic pattern: geographic patterns:
- Top 10 departments: 18-20 types (metropolitan france)
- bottom 10 departments: 1-11 types (overseas territories)
- Departments 984, 986, 987: 1-3 types only
  --> French overseas territories havd limited cultural infrastructure
  
#### Statistical conclusion: 
We reject the null hypothesis of uniform diversity distribution (pearson r = 0.73, p < 0.001). 
Cultural diversity is significantly correlated with the total number of venues.
--> Larger cultural ecosystems support greater variety of cultural offerings.

### Research question 5 : PACA regional analysis
#### How does PACA compare to national metrics?


```python

# research question 5: paca regional analysis

#Identify paca departments


# paca departments (provence-alpes-côte d'azur)
paca_departments = ['04', '05', '06', '13', '83', '84']

# create paca flag based on department code
df['is_paca'] = df['dept_code'].isin(paca_departments)

print(f"\npaca departments: {', '.join(paca_departments)}")
print(f"   04 = alpes-de-haute-provence")
print(f"   05 = hautes-alpes")
print(f"   06 = alpes-maritimes")
print(f"   13 = bouches-du-rhône")
print(f"   83 = var")
print(f"   84 = vaucluse")


# Basic statistics


paca_venues = df[df['is_paca']]
other_venues = df[~df['is_paca']]


print(f"\nStatistics PACA:")
print(f"   venues: {len(paca_venues):,} ({len(paca_venues)/len(df)*100:.1f}%)")
print(f"   communes: {paca_venues['code_insee'].nunique():,}")
print(f"   departments: {paca_venues['dept_code'].nunique()}")

print(f"\nrest of france:")
print(f"   venues: {len(other_venues):,} ({len(other_venues)/len(df)*100:.1f}%)")
print(f"   communes: {other_venues['code_insee'].nunique():,}")
print(f"   departments: {other_venues['dept_code'].nunique()}")
```

```python
# Aggregate density by unique communes


# paca: group by commune first to get unique population
paca_by_commune = paca_venues.groupby('code_insee').agg({
    'id': 'count',
    'population': 'first'
}).rename(columns={'id': 'venues'})

paca_by_commune = paca_by_commune[paca_by_commune['population'].notna()]
paca_by_commune = paca_by_commune[paca_by_commune['population'] > 0]

paca_total_venues = paca_by_commune['venues'].sum()
paca_total_population = paca_by_commune['population'].sum()
paca_aggregate_density = (paca_total_venues / paca_total_population) * 100000

print(f"\nPACA:")
print(f"Communes: {len(paca_by_commune):,}")
print(f"Total venues: {paca_total_venues:,}")
print(f"Total population: {paca_total_population:,.0f}")
print(f"Aggregate density: {paca_aggregate_density:.2f} per 100k")
print()
# Rest of france: group by commune first
other_by_commune = other_venues.groupby('code_insee').agg({
    'id': 'count',
    'population': 'first'
}).rename(columns={'id': 'venues'})

other_by_commune = other_by_commune[other_by_commune['population'].notna()]
other_by_commune = other_by_commune[other_by_commune['population'] > 0]

other_total_venues = other_by_commune['venues'].sum()
other_total_population = other_by_commune['population'].sum()
other_aggregate_density = (other_total_venues / other_total_population) * 100000

print(f"\nRest of france:")
print(f"   communes: {len(other_by_commune):,}")
print(f"   total venues: {other_total_venues:,}")
print(f"   total population: {other_total_population:,.0f}")
print(f"   aggregate density: {other_aggregate_density:.2f} per 100k")
print()
# comparison
ratio = paca_aggregate_density / other_aggregate_density
diff_pct = ((paca_aggregate_density - other_aggregate_density) / other_aggregate_density) * 100

print()
print(f"Comparaison :ratio: {ratio:.2f}")
if ratio > 1:
    print(f"PACA has {diff_pct:.1f}% higher density than rest of france")
else:
    print(f"PACA has {abs(diff_pct):.1f}% lower density than rest of france")
```

```python
# Statistical test (t-test on commune densities)


# calculate density per commune for both groups
paca_by_commune['density'] = (paca_by_commune['venues'] / paca_by_commune['population']) * 100000
other_by_commune['density'] = (other_by_commune['venues'] / other_by_commune['population']) * 100000

# remove inf/nan
paca_densities = paca_by_commune['density'].replace([np.inf, -np.inf], np.nan).dropna()
other_densities = other_by_commune['density'].replace([np.inf, -np.inf], np.nan).dropna()

print("\nHypotheses:")
print("H0: No difference in cultural density between paca and rest of france")
print("H1: Significant difference exists")


# perform t-test
t_stat, p_value = st.ttest_ind(paca_densities, other_densities)

print(f"\nT-test:")
print(f"T-statistic: {t_stat:.4f}")
print(f"P-value: {p_value:.10f}")
print(f"PACA sample: {len(paca_densities):,} communes")
print(f"Other sample: {len(other_densities):,} communes")

if p_value < 0.05:
    print(f"\n   reject h0 (p < 0.05)")
    if paca_aggregate_density < other_aggregate_density:
        print(f"PACA has significantly lower cultural density")
    else:
        print(f"PACa has significantly higher cultural density")
else:
    print(f"\nFail to reject h0 (p >= 0.05)")
   
    
    # No significant difference
```

```python
# Descriptive statistics: comparaison PACA communes / rest of France

print(f"\nPACA communes:")
print(f"Mean: {paca_densities.mean():.2f} per 100k")
print(f"Median: {paca_densities.median():.2f} per 100k")
print(f"Std: {paca_densities.std():.2f}")

print(f"\nRest of France communes:")
print(f"Mean: {other_densities.mean():.2f} per 100k")
print(f"Median: {other_densities.median():.2f} per 100k")
print(f"Std: {other_densities.std():.2f}")
```

```python
#  Bar chart comparison


fig, ax = plt.subplots(figsize=(10, 6))

categories = ['paca', 'rest of france']
densities = [paca_aggregate_density, other_aggregate_density]
colors = ['#FF6F61', '#26A69A']

bars = ax.bar(categories, densities, color=colors, edgecolor='#004D40', linewidth=1.5)

# add value labels
for i, (cat, dens) in enumerate(zip(categories, densities)):
    ax.text(i, dens + 3, f'{dens:.1f}', ha='center', va='bottom', fontweight='bold')

ax.set_ylabel('aggregate density (venues per 100k inhabitants)', fontweight='bold')
ax.set_title('cultural density: paca vs rest of france', fontweight='bold', fontsize=14)
ax.set_ylim(0, max(densities) * 1.15)
ax.grid(axis='y', alpha=0.3)

plt.tight_layout()
plt.show()

```

#### Key finding

PACA has 28.5% lower cultural density than the rest of france:
- PACA: 94.7 venues per 100k inhabitants
- Rest of France: 132.4 venues per 100k inhabitants.
--> This means that for every 100,000 people, paca has 38 fewer cultural venues than the national average.

**Scale of the difference**
PACA region (5.2 million inhabitants):
- Actual venues: 4,899
- If paca had national average density: 6,897 venues
- Missing venues: ~2,000 cultural facilities

**Statistical interpretation : aggregate density (what matters for access):**
- Clear difference: paca has 28.5% lower density.
- This is a real, measurable gap in cultural infrastructure.
Meaning: residents have fewer cultural options by head.

**t-test on commune distribution:**
p = 0.48 (not significant)
Meaning: the pattern of distribution is similar : both PACA and rest of France have similar types of communes (some rich in culture, some poor).

So:
--> PACA doesn't have a different distribution pattern, it just has less cultural venues overall : the pattern is the same (layout), but the quantity is lower (fewer products).
--> PACA follows national patterns (urban areas have more venues, rural areas have less) but at a lower baseline.

**Possible explanations:**

1. Geographic factors:
- Mountainous terrain (Alps) limits infrastructure development
- Dispersed population in alpine departments (04, 05)
- Coastal concentration vs. rural interior (Cote d'Azur vs "Provence Verte")

2. Demographic factors:
- Tourism economy may prioritize different infrastructure
- Seasonal population variations
- Recent urbanization patterns

3. Economic factors:
- Lowercultural budget (by person)
- Competition with other infrastructure priorities
- Historical investment patterns

**Conclusion:**
PACA region shows significantly lower cultural density (-28.5%) compared to the french national average, representing approximately 2,000 missing cultural facilities across the region. 
While the distribution pattern across communes mirrors national trends (urban concentration, rural scarcity), the overall baseline is lower. 
This suggests systemic underinvestment in cultural infrastructure rather than unique distribution problems.
