CODE 4: REST API (FastAPI)
Purpose
Create REST API with 5 endpoints to expose cultural data programmatically


I implemented a REST API because my project data is stored in MySQL (communes, cultural venues, socio-economic indicators) and I needed a reproducible way to query it from different clients (Swagger tests, a future dashboard, or analysis scripts) without direct database access. The API acts as a controlled interface: it standardizes inputs (query parameters), outputs (JSON), and error handling (HTTP status codes), which makes results easier to verify and document in a report.

I chose REST because the project fits a “resources + queries” model: communes and venues are resources, and the main operations are read/list/filter plus a few derived statistics. REST over HTTP is simple to test, widely understood, and compatible with typical web tooling.

For the grading requirement (“5 endpoints excluding root/health”), the business endpoints are:

GET /venues — returns cultural venues with filters (commune code, venue type) and pagination (limit/offset).

GET /communes — lists communes with filters (department/region) and name search; needed to explore territories before running stats.

GET /stats/commune/{commune_code} — produces a commune profile by joining venues + population + income data (count, venues per 1,000 inhabitants, socio-economic context).

GET /stats/density — ranks communes by venues per 1,000 inhabitants (with a minimum population threshold) to identify high/low cultural supply.

GET /stats/dept/{dept_code} (or /stats/region/{region_code}) — aggregates the same indicators at a larger scale to compare territories and avoid relying only on single-commune views.

Root (/) and health (/health) exist for navigation and technical verification, but they are not counted as “project endpoints.” Swagger (/docs) is used to demonstrate each endpoint with request/response evidence (parameters, status codes, JSON output) through screenshots.
